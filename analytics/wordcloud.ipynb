```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np # linear algebra
import nltk
from nltk.corpus import stopwords
from nltk.corpus import words
from nltk.tokenize import word_tokenize

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
nltk.download('words')
```

```python
words = words.words()
```

```python
blacklist = ["html", "http", "https", "www", "gt", "amp", "com",
"nbsp", "em" , "en" , "lt" , "quot" , "chatgpt" , "gpt"]
```

```python
df = pd.read_csv("../Temp/sentiment.csv")
```

# All

```python
all_df = df['clean_text']
all_df_whitelist = [word for word in all_df if word not in set(STOPWORDS) and word.strip() not in blacklist]
```

```python
len(all_df_whitelist)
```

```python
from collections import Counter
Counter = Counter(all_df_whitelist) 
```

```python
most_occur_all = Counter.most_common(5000) 
most_occur_all[0:10]
```

```python
all_filter = []
for sent in most_occur_all:
    words = sent[0].split()
    for word in words:
        if word.strip() not in blacklist:
            all_filter.append(word.strip())
all_filter
```

```python
all_text = ' '.join(all_filter)
```

```python
all_wordcloud = WordCloud(font_step=2, width=800, height=400,
                stopwords = set(STOPWORDS), background_color = 'white').generate(all_text)
```

```python
# Display the generated image:
plt.figure(figsize=(10,8))
plt.imshow(all_wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig(f'../Temp/all.jpg', format='jpg')
plt.show()
```

# Monthly

```python
df_monthly = df.copy()
```

```python
df_monthly["datetime"] = pd.to_datetime(df_monthly["datetime"])
```

```python
monthly_df = df_monthly.groupby(pd.Grouper(key='datetime', freq='M'))
```

```python
for month, per_df in monthly_df:
    # Concatenate text from 'clean_text' column for each month
    month_df = per_df['clean_text']
    whitelist = [word for word in month_df if word not in set(STOPWORDS) and word.strip() not in blacklist]
    from collections import Counter
    Counter = Counter(whitelist) 
    most_occur = Counter.most_common(5000) 

    words_filter = []
    for sent in most_occur:
        words = sent[0].split()
        for word in words:
            if word.strip() not in blacklist and len(word) > 1:
                words_filter.append(word.strip())
    text = ' '.join(words_filter)
    
    wordcloud = WordCloud(font_step=2, width=800, height=400,
                stopwords = set(STOPWORDS), background_color = 'white').generate(text)
    
    # Plot and save word cloud for the month
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word cloud for {month}')
    plt.savefig(f'../Temp/wordcloud_{month}_v1.jpg', format='jpg')
    plt.show()
```
